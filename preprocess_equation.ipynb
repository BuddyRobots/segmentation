```python
import cv2
import os
from matplotlib import pyplot as plt
%matplotlib inline
input_dir = "raw_equations"
output_dir = "binarized_equations"
```

```python
for (dirpath, dirnames, filenames) in os.walk(input_dir):
#     print(filenames)
    for filename in filenames:
        image = cv2.imread(input_dir + "/" + filename, 0)
        binary_image = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 21)
        cv2.imwrite(output_dir + "/" + filename.replace("jpg", "png"), binary_image)
#         break
```
